```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Visualization and its analysis

```python
# load the final .csv file
df = pd.read_csv('final_df.csv')
```

```python
df.shape
```

Afte loading the final .csv file, which has 1666 rows after removing retweets

I am interested to look into types of dog names are available

```python
# 
df.name.value_counts()[:10].sort_values(ascending=True).plot.barh()
plt.title( 'Name and their count')
plt.xlabel('Count');
```

There were many of them are none and strange names. Cooper tops the list

Next, the raltion between retweets count and favourites i.e. likes for the post i have looked into.

```python
# retweets and followers
cols = ['retweet_count','favorite_count', 'rating_numerator', 'rating_denominator' ]
cm = df[cols].corr()
plt.figure(figsize=(15, 7))
sns.heatmap(cm);
```

Retweet and favourites are higly correlated. rating numerator and denomminator with retweets and favourites are also correlted. 

```python
sns.regplot(x= 'retweet_count', y = 'favorite_count', data = df)
plt.title("Favorites and Retweets", weight = 'bold')
plt.xlabel('Retweets_count')
plt.ylabel('Favorites_count')
```

It shows that retweets an favourites are highly correlated. i can be infer that who ever retweeted have done with like of the post and shows that most of the retweets happen if they like the post
